```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html"
html_data  = requests.get(url).text
html_data[:100]
```

# Extracting Data using BeautifulSoup

```python
soup = BeautifulSoup(html_data, 'html5lib')

soup.title.text
```

```python
amazon_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"])
amazon_data
```

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text

    new_row={"Date":date, "Open":Open, "High":high, "Low":low, "Close":close, "Adj Close":adj_close, "Volume":volume}
    amazon_data = pd.concat([amazon_data, pd.DataFrame(new_row, index=[0])], ignore_index=True)
```
